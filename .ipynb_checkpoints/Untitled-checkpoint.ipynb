```python
import pandas as pd
import numpy as np
```

```python
movie_df = pd.read_csv("C:/Users/saads/Downloads/Netflix TV Shows and Movies.csv")
movie_df
```

```python
movie_renamed = movie_df.rename(index=str, columns={"runtime":"movie_length_min"})
movie_renamed.info
```

```python
movie_renamed.isnull().sum()
```

```python
movie_renamed["description"] = movie_renamed["description"].fillna("No Information available")
```

```python
#better option
#from sklearn.impute import KNNImputer

#imputer = KNNImputer(n_neighbors=5)
#df[['imdb_score', 'imdb_votes', 'movie_length_min']] = imputer.fit_transform(
 #   df[['imdb_score', 'imdb_votes', 'movie_length_min']])

movie_median =  movie_renamed["imdb_votes"].median()
```

```python
movie_renamed["imdb_votes"] = movie_renamed["imdb_votes"].fillna(movie_median)
```

```python
movie_renamed.head()
```

```python
movie_renamed.isnull().sum()
```

```python
R=movie_renamed["imdb_score"]
v=movie_renamed["imdb_votes"]
m=movie_renamed["imdb_votes"].quantile(0.70)
C=movie_renamed["imdb_score"].mean()
```

```python
movie_renamed['weighted_Average'] =((R*v)+ (C*m))/(v+m)
```

```python
movie_renamed.head()
```

```python
movie_ranking_sorted=movie_renamed.sort_values('weighted_Average', ascending=False)
movie_ranking_sorted[['title','release_year','movie_length_min','imdb_score','imdb_votes','weighted_Average']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=movie_ranking_sorted.sort_values('weighted_Average', ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_Average'].head(10), y=weight_average['title'].head(10))
plt.xlim(3,10)
plt.title('Best Movies by average votes')
plt.xlabel('weighted average score')
plt.ylabel('Movie title')
plt.savefig('best_movies.png')
```

```python
plt.figure(figsize=(12,6))
axis2=sns.scatterplot(x='weighted_Average', y='movie_length_min', data=weight_average, hue='movie_length_min', palette='viridis' )
plt.xlim(3,10)
plt.title('Best Movies by average votes')
plt.xlabel('weighted average score')
plt.ylabel('movie_length_min')
plt.savefig('best_movies.png')
```

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
columns_to_scale = ['weighted_Average', 'movie_length_min']

movie_scaled_df=scaling.fit_transform(movie_renamed[columns_to_scale])

movie_normalized_df= pd.DataFrame(movie_scaled_df,columns=['normalized_weight_average', 'normalized_movie_length'])

movie_normalized_df.index = movie_renamed.index
movie_normalized_df.head()
```

```python
movie_normalized_df.isnull().sum()
```

```python
movie_renamed[['normalized_weight_average', 'normalized_movie_length']] = movie_normalized_df
```

```python
movie_renamed.head()
```

```python
movie_renamed['score'] = movie_renamed['normalized_weight_average']*0.7 + movie_renamed['normalized_movie_length']*0.3
movie_scored_df =  movie_renamed.sort_values(['score'],ascending=False)
movie_scored_df.head()
```

```python

plt.figure(figsize=(12,6))
axis3= sns.barplot(x='score', y='title' , data=movie_scored_df.head(10))
plt.xlim(0.4,0.9)
plt.title('Best Movies by average votes')
plt.xlabel('score')
plt.ylabel('Movie title')
plt.savefig('best_movies.png')
plt.show()
```
